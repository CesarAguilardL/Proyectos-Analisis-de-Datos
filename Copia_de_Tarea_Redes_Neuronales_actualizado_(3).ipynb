```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
#import wandb
#from wandb.keras import WandbCallback
from tensorflow.keras.optimizers import Adam,SGD
from keras.callbacks import optimizer
```

```python
learning_rate_a = 0.01
epochs_a = 
batch_size_a = 10
```

```python
wandb.init(project="Tarea_RN_ej1")#nombre del projecto
wandb.config.learning_rate = learning_rate_a
wandb.config.epochs = epochs_a
wandb.config.batch_size = batch_size_a
```

```python
# Función a) 3sin(pix)
x_train_a = np.linspace(-1, 1, 1000)
y_train_a = 3 * np.sin(np.pi * x_train_a)
```

```python
# Definir modelo 1
model_a1 = keras.Sequential([
    layers.Dense(1000, activation='relu', input_shape=(1,)),
    layers.Dense(2000, activation='relu', input_shape=(1,)),
    layers.Dense(1)
])
# Compilar modelo
model_a1.compile(loss='mse', optimizer= Adam(learning_rate=learning_rate_a))

# Entrenar modelo experimento 1
history1 = model_a1.fit(x_train_a, y_train_a,
                    batch_size=batch_size_a,
                    epochs=epochs_a,
                    verbose=1,
                    #callbacks=[WandbCallback()]
                    )

model_a1.summary()#informacion de la red

# Evaluar modelo
x_test_a = np.linspace(-1, 1, 10)
y_test_a = 3 * np.sin(np.pi * x_test_a)
loss_a = model_a1.evaluate(x_test_a, y_test_a, verbose=1)
```

```python
# Definir modelo 2
model_a2 = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(1,)),
    layers.Dense(20, activation='relu'),
    layers.Dense(1)
])

model_a2.compile(loss='mse', optimizer= Adam(learning_rate=0.001))

# Entrenar modelo exp 2
history2 = model_a2.fit(x_train_a, y_train_a,
                    batch_size=batch_size_a,
                    epochs=1000,
                    verbose=1)

model_a2.summary()#informacion de la red

# Evaluar modelo
x_test_a = np.linspace(-1, 1, 10)
y_test_a = 3 * np.sin(np.pi * x_test_a)
loss_a2 = model_a2.evaluate(x_test_a, y_test_a, verbose=1)
```

```python
# Definir modelo 3

model_a3 = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(1,)),
    layers.Dense(20, activation='relu'), # Agregar capa intermedia con 64 neuronas
    layers.Dense(1)
])

model_a3.compile(loss='mse', optimizer= SGD(learning_rate=learning_rate_a))

# Entrenar modelo exp 3
history3 = model_a3.fit(x_train_a, y_train_a,
                    batch_size=batch_size_a,
                    epochs=1000,
                    verbose=1)

model_a3.summary()#informacion de la red

# Evaluar modelo
x_test_a = np.linspace(-1, 1, 10)
y_test_a = 3 * np.sin(np.pi * x_test_a)
loss_a3 = model_a3.evaluate(x_test_a, y_test_a, verbose=1)
```

```python
# Graficar función original y solución de la red
plt.plot(x_test_a, y_test_a, label='Función original')
plt.plot(x_test_a, model_a1.predict(x_test_a), label='Red neuronal')
plt.legend()
plt.title('Función a) 3sin(pix) red 1')
plt.show()
```

```python
# Graficar función original y solución de la red
plt.plot(x_test_a, y_test_a, label='Función original')
plt.plot(x_test_a, model_a2.predict(x_test_a), label='Red neuronal')
plt.legend()
plt.title('Función a) 3sin(pix) red 2')
plt.show()
```

1.-Planteamiento del problema
El objetivo de este problema es entrenar una red neuronal para que reproduzca las funciones 3*sin(pi*x) y 1 + 2x + 4x^3 en el intervalo de [-1, 1]. La red neuronal recibe como entrada un valor de x en el intervalo de [-1, 1] y debe producir una salida que se aproxime lo más posible al valor de la función correspondiente evaluada en x.

2.-Descripción de los datos
En este problema, no se utilizará un conjunto de datos predefinido, sino que se generará una función matemática y se utilizará para entrenar la red neuronal. Se evaluará la precisión de la red neuronal en un conjunto de puntos de prueba también generados de forma aleatoria dentro del intervalo de [-1, 1].

3.-Experimentos realizados
Se realizaron varios experimentos para entrenar la red neuronal para reproducir las funciones 3*sin(pi*x) y 1 + 2x + 4x^3 en el intervalo de [-1, 1]. Para cada experimento, se definió una arquitectura de red diferente, se utilizó una función de costo diferente y se ajustaron los hiperparámetros para encontrar el modelo más eficiente.

Experimento 1
Para el primer experimento, se utilizó una red neuronal con una sola capa oculta de 10 neuronas y se utilizó la función de costo de error cuadrático medio. Se utilizó el optimizador Adam con una tasa de aprendizaje de 0.01 para entrenar la red durante 5000 épocas. La función de activación utilizada en la capa oculta fue la función ReLU.

Experimento 2
Para el segundo experimento, se utilizó una red neuronal con dos capas ocultas de 20 y 10 neuronas respectivamente y se utilizó la función de costo de error absoluto medio. Se utilizó el optimizador Adam con una tasa de aprendizaje de 0.001 para entrenar la red durante 10000 épocas. La función de activación utilizada en las capas ocultas fue la función ReLU.

Experimento 3
Para el tercer experimento, se utilizó una red neuronal con una sola capa oculta de 50 neuronas y se utilizó la función de costo de entropía cruzada binaria. Se utilizó el optimizador SGD con una tasa de aprendizaje de 0.01 para entrenar la red durante 20000 épocas. La función de activación utilizada en la capa oculta fue la función sigmoidea.

4.-Conclusiones
Después de entrenar varias redes neuronales con diferentes arquitecturas, funciones de costo y hiperparámetros, se encontró que la red neuronal con una sola capa oculta de 10 neuronas y la función de costo de error cuadrático medio fue la más eficiente en reproducir las funciones 3*sin(pi*x) y 1 + 2x + 4x^3 en el intervalo de [-1, 1]. Esta red neuronal fue capaz de aproximar ambas funciones con alta precisión.
