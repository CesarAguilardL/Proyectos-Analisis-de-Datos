```python
# Funci√≥n b) 1+2x+4x^3
x_train = np.linspace(-1, 1, 1000)
y_train = 1 + 2 * x_train + 4 * x_train**3
```

```python
# Definir modelo
model_b = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(1,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])
```

```python
# Compilar modelo
model_b.compile(loss='mse', optimizer='adam')

# Entrenar modelo
history_b = model_b.fit(x_train, y_train, epochs=500, verbose=0)
```

```python
# Evaluar modelo
x_test = np.linspace(-1, 1, 100)
y_test = 1 + 2 * x_test + 4 * x_test**3
loss_b = model_b.evaluate(x_test, y_test, verbose=0)
```

```python
# Graficar funci√≥n original y soluci√≥n de la red
plt.plot(x_test, y_test, label='Funci√≥n original')
plt.plot(x_test, model_b.predict(x_test), label='Red neuronal')
plt.legend()
plt.title('Funci√≥n b) 1+2x+4x^3')
plt.show()
```
